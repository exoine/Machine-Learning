```python
import pandas as pd
```

```python
data=pd.read_csv(r'C:\Users\SAMIA DAWOOD\OneDrive\Desktop\atp_matches_2000.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
tourney_name=LabelEncoder()
surface=LabelEncoder()
tourney_level=LabelEncoder()
winner_entry=LabelEncoder()
```

```python
data['tourney_name']=tourney_name.fit_transform(data['tourney_name'])
data['surface']=surface.fit_transform(data['surface'])
data['tourney_level']=tourney_level.fit_transform(data['tourney_level'])
data['winner_entry']=winner_entry.fit_transform(data['winner_entry'])
```

```python
data
```

```python
features_cols=['tourney_name','surface','tourney_level']
x=data[features_cols]
y=data.winner_entry
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier=DecisionTreeClassifier(criterion='gini')
```

```python
classifier.fit(x_train,y_train)
```

```python
classifier.predict(x_test)
```
