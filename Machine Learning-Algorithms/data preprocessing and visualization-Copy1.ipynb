```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#loading the stock data from csv file
data=pd.read_csv('stocks.csv')
```

```python
#for visualizing the closing price of the stock over time, first plot a line plot using seaborn
plt.figure(figsize=(10,6))
sns.lineplot(x='Date', y='Close', data=data)
plt.xlabel('DATE')
plt.ylabel('CLOSING PRICE')
plt.title('CLOSING PRICE OF TOCK OVER TIME')
plt.xticks(rotation=45)
plt.show()
```

```python
#now for relationship between tock volume and the closing price we use the scatter plot from matplotlib
plt.figure(figsize=(10,6))
plt.scatter(data['Volume'], data['Close'], color='black',alpha=0.5)
plt.xlabel('Volume')
plt.ylabel('CLOSING PRICE')
plt.title('RELATIONSHIP B\W STOCK VOLUME AND CLOSING PRICE')
plt.show()
```

```python
#to display the distribution of open, high, low and close prices of the Stock we use the boxplot fromseaborn library
plt.figure(figsize=(10,6))
sns.boxplot(data=data[['Open','High','Low','Close']])
plt.xlabel('PRICE')
plt.ylabel('DISTRIBUTION')
plt.title('STOCK PRICE DISTRIBUTION')
plt.show()
```
