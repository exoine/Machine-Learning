## Applying Standardization

```python
# Selecting the columns of interest
cols_of_interest = ['Electrodermal activity', 'Time interval']
df_selected = df[cols_of_interest]
```

```python
# Standardize the selected columns
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_selected)
```

```python
# Converting the scaled data to a dataframe
df_scaled = pd.DataFrame(df_scaled, columns=cols_of_interest)
```

```python
# Converting the scaled data to a dataframe
df_scaled = pd.DataFrame(df_scaled, columns=cols_of_interest)
```

```python
# Replacing the original columns with the standardized columns in the original dataframe
df[cols_of_interest] = df_scaled
```

### Let's create a function that converts all negative values in a particular column of a dataset to their positive equivalents.

```python
def make_positive(x):
    if x < 0:
        return abs(x)
    else:
        return x
```

```python
df['Blood Value pulse'] = df['Blood Value pulse'].apply(make_positive)
df['Electrodermal activity'] = df['Electrodermal activity'].apply(make_positive)
df['Heart rate'] = df['Heart rate'].apply(make_positive)
df['Time interval'] = df['Time interval'].apply(make_positive)
df['Temperature'] = df['Temperature'].apply(make_positive)
```

```python
df.tail(30)
```

### Now that we have preprocessed the dataset, we will use heart rate data to predict the level of stress in the individuals.

### Below we will make a fucntion which is predicting stress level using heart rate

* If a person's heart rate is above 100, it is an indication that they may be experiencing some level of stress. On the other hand, if their heart rate is below 100, it suggests that the person is not currently experiencing stress.

* Reference for above mentioned comment.
www.verywellhealth.com/hospital-rapid-heart-rate-5216290 

```python
def transform_heart_rate(x):
    if x > 100:
        return 1
    else:
        return 0
```

```python
df['Stress level'] = df['Heart rate'].apply(transform_heart_rate)
```

```python
df.head()
```

### We will now create a plot to visualize the relationship between Heart rate and Stress level.

```python
sns.set(style="ticks", color_codes=True)
sns.pairplot(df, x_vars=['Heart rate'], y_vars=['Stress level'], height=4)
plt.show()
```

* The graph above indicates that nurses are more prone to stress compared to other professions, as evident from the high number of nurses with a heart rate above 0 and 1 on the Y-axis. This suggests that a majority of nurses may be experiencing stress, as their heart rate is higher. To further analyze this trend, we can use a linear regression model to make predictions based on the data.

```python
# Select the columns of interest
cols_of_interest = ['Heart rate', 'Blood Value pulse', 'Temperature']
df_selected = df[cols_of_interest]

# Create a pairplot with density plots on the diagonal
sns.pairplot(data=df_selected, diag_kind='kde')
```

## Importing the Libraries to Build a model for Prediction

* We are using LinearRegression for it

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df.drop('Stress level', axis=1), df['Stress level'], test_size=0.2)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
print(y_pred)
```

```python
score = r2_score(y_test, y_pred)
```

```python
print('R-squared score:', score)
```

```python
mse = mean_squared_error(y_test, y_pred)
```

```python
print('Mean squared error:', mse)
```

```python
# Plot the scatter with predicted values in red color
plt.scatter(X_test['Heart rate'], y_test, color='red', label='Actual Stress level')
plt.scatter(X_test['Heart rate'], y_pred, color='green', label='Predicted Stress level')
plt.xlabel('Heart rate')
plt.ylabel('Stress level')
plt.title(f'Linear Regression (R2 score: {score:.2f})')
plt.legend(loc='upper left')
plt.show()
```
